```python
state_list = ['Andhra_Pradesh','Arunachal_Pradesh','Assam','Bihar','Chhattisgarh','Goa','Gujarat','Haryana','Himachal_Pradesh','Jharkhand','Karnataka','Kerala','Madhya_Pradesh','Maharashtra','Manipur','Meghalaya'
              ,'Mizoram','Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil_Nadu','Telangana','Tripura','Uttar_Pradesh','Uttarakhand','West_Bengal','Andaman_and_Nicobar_Islands','Chandigarh'
              ,'Dadra_and_Nagar_Haveli_and_Daman_and_Diu','Jammu_and_Kashmir','Ladakh','Lakshadweep','National_Capital_Territory_of_Delhi','Puducherry','Other_Territory']

df_cols = ['State', 'Apr-21', 'Apr-20', 'Apr-19', 'Apr-18', 'May-21', 'May-20',
       'May-19', 'May-18', 'Jun-21', 'Jun-20', 'Jun-19', 'Jun-18', 'Jul-21',
       'Jul-20', 'Jul-19', 'Jul-18', 'Aug-21', 'Aug-20', 'Aug-19', 'Aug-18',
       'Sep-21', 'Sep-20', 'Sep-19', 'Sep-18', 'Oct-21', 'Oct-20', 'Oct-19',
       'Oct-18', 'Nov-21', 'Nov-20', 'Nov-19', 'Nov-18', 'Dec-21', 'Dec-20',
       'Dec-19', 'Dec-18', 'Jan-21', 'Jan-20', 'Jan-19', 'Jan-18', 'Feb-21',
       'Feb-20', 'Feb-19', 'Feb-18', 'Mar-21', 'Mar-20', 'Mar-19', 'Mar-18']
df = pd.DataFrame(columns = df_cols)
```

```python
for i,state in zip(range(1,40),state_list):
    cur_data = data[i].droplevel(level = 0, axis = 1)
    cur_data.drop('M-O-M Change', axis = 1,inplace = True)
    cur_data = cur_data.apply(lambda x: x.replace({'₹':''},regex=True))
    cur_data = cur_data.astype(str).applymap(lambda x: x.split()[0])
    cur_data.columns = ['Month','21','20','19','18']
    index_lst = ['Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Jan','Feb','Mar']
    cur_data['Month'] = index_lst
    cur_data.set_index('Month', inplace = True)
    cur_data = cur_data.stack().to_frame().T
    cur_data.columns = ['{}-{}'.format(*c) for c in cur_data.columns]
    cur_data.insert(loc=0, column='State', value=state)
    row_dict = dict(zip(df_cols, cur_data.iloc[0].values.tolist()))
    df = df.append(row_dict,ignore_index = True)
```

```python
df.to_csv('wiki_data.csv', index=False)
```
